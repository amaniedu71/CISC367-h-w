## Data Acquisition & Preparation

The data was retrieved from [www.fueleconomy.gov](www.fueleconomy.gov), the official U.S government source for fuel economy information for most models of cars. The data is comprehensive and includes fuel MPG data from 1998 to present (2021). 

Moreover, they provide a developer friendly API for querying their data as well as a cleaned up CSV file which is downloaded below. Find more information about the data here: [fueleconomy.gov web services](https://www.fueleconomy.gov/feg/ws/index.shtml)

```python
cars = pd.read_csv("https://www.fueleconomy.gov/feg/epadata/vehicles.csv")
cars.rename_axis("entries") #rename key to be called entries
```

Here we take a glimpse of the data. 

```python
print(cars.describe()) #describe dataframe
print(cars.shape) #rows x cols
print(cars.columns) #unique column labels
print(cars.head(3)) #slice first 3 rows
```

Here we select the columsn that we want to keep. 
We select only 19 columns out of the 80 in order to be friendly for use. 
These are the selected columns: 

* drive - drive axle type
* cylinders - engine cylinders
* trany - transmission
* city08 - city MPG for fuelType1
* cityA08 - city MPG for fuelType2
* cityE - city electricity consumption in kw-hrs/100 miles
* fuelType1 - fuel type 1. For single fuel vehicles, this will be the only fuel. For dual fuel vehicles, this will be the conventional fuel.
* fuelType2 - fuel type 2. For dual fuel vehicles, this will be the alternative fuel (e.g. E85, Electricity, CNG, LPG). For single fuel vehicles, this field is not used
* highway08 - highway MPG for fuelType1 
* highwayA08 - highway MPG for fuelType2
* highwayE - highway electricity consumption in kw-hrs/100 miles
* make - manufacturer (division)
* model - model name (carline)
* year - model year
* eng_dscr - engine descriptor; see [list of descriptors](http://www.fueleconomy.gov/feg/findacarhelp.shtml#engine)
* VClass - EPA vehicle size class [List of classes](https://en.wikipedia.org/wiki/Vehicle_size_class)
* lv4 - 4 door luggage volume (cubic feet) 
* lv2 - 2 door luggage volume (cubic feet) 
* hlv - hatchback luggage volume (cubic feet)

**special note** 

* Two fueltypes are included since cars sometimes have two different fuels eg. electricity and petrol

```python
cars_final = cars[["drive", "cylinders", "trany", "city08", "cityA08", "cityE", "highway08", "highwayA08", "highwayE",  "make", "model", "year", "eng_dscr", "VClass", "lv4", "lv2", "hlv", "fuelType1", "fuelType2"]]
```

## Convert to CSV

Here we convert the dataframe to csv using the to_csv() method. In order to match format of CORGIS data, we specify to not include headers or index. 

```python
cars_final.to_csv('cars-corgis.csv', header = False, index = False)
```
